```python
from pyspark.sql import functions as F
from pyspark.sql import types as T

ATLAS_TABLES = [
    "BPC_Plug_Financials",
    "BPC_Plug_Financials_MachineCosts",
    "BW_COPA_DailyFeePoker_Units",
    "BW_ZRBINQ_BillingPlans",
    "GameOpsInstallBaseActivity",
    "GameOpsInstallBasePSD",
    "RPM_Exp_Depreciation",
    "RPM_Intl_Performance_EMEA_Africa",
    "RPM_Intl_Performance_EMEA_FixedFee",
    "RPM_Intl_Performance_EMEA_GreeceWLA",
    "RPM_Intl_Performance_LAC",
    "RPM_Peformance_PremLotto_DE",
    "RPM_Peformance_PremLotto_NY",
    "RPM_Peformance_PremLotto_RI",
    "RPM_Performance_CDS_Class_II",
    "RPM_Performance_CDS_Class_II_ManualBilling",
    "RPM_Performance_WAP_NewCanada",
]

SILVER_DDL = {}

SILVER_DDL["BPC_Plug_Financials"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.BPC_Plug_Financials (
    unnamed_1           STRING,
    jan                 DOUBLE,
    feb                 DOUBLE,
    mar                 DOUBLE,
    apr                 DOUBLE,
    may                 DOUBLE,
    jun                 DOUBLE,
    jul                 DOUBLE,
    aug                 DOUBLE,
    sep                 DOUBLE,
    oct                 DOUBLE,
    nov                 DOUBLE,
    dec                 DOUBLE,
    jan_1               DOUBLE,
    feb_1               DOUBLE,
    mar_1               DOUBLE,
    apr_1               DOUBLE,
    may_1               DOUBLE,
    jun_1               DOUBLE,
    jul_1               DOUBLE,
    aug_1               DOUBLE,
    sep_1               DOUBLE,
    oct_1               DOUBLE,
    nov_1               DOUBLE,
    dec_1               DOUBLE,
    jan_2               DOUBLE,
    feb_2               DOUBLE,
    mar_2               DOUBLE,
    apr_2               DOUBLE,
    may_2               DOUBLE,
    jun_2               DOUBLE,
    jul_2               DOUBLE,
    aug_2               DOUBLE,
    sep_2               DOUBLE,
    oct_2               DOUBLE,
    nov_2               DOUBLE,
    dec_2               DOUBLE,
    jan_3               DOUBLE,
    feb_3               DOUBLE,
    mar_3               DOUBLE,
    apr_3               DOUBLE,
    may_3               DOUBLE,
    june                DOUBLE,
    july                DOUBLE,
    august              DOUBLE,
    september           DOUBLE,
    october             DOUBLE,
    november            DOUBLE,
    december            DOUBLE,
    january             DOUBLE,
    february            DOUBLE,
    march               DOUBLE,
    april               DOUBLE,
    may_4               DOUBLE,
    june_1              DOUBLE,
    july_1              DOUBLE,
    august_1            DOUBLE,
    september_1         DOUBLE,
    october_1           DOUBLE,
    november_1          DOUBLE,
    december_1          DOUBLE,
    january_1           DOUBLE,
    february_1          DOUBLE,
    march_1             DOUBLE,
    april_1             DOUBLE,
    may_5               DOUBLE,
    june_2              DOUBLE,
    july_2              DOUBLE,
    august_2            DOUBLE,
    september_2         DOUBLE,
    october_2           DOUBLE,
    november_2          DOUBLE,
    december_2          DOUBLE,
    january_2           DOUBLE,
    february_2          DOUBLE,
    march_2             DOUBLE,
    april_2             DOUBLE,
    may_6               DOUBLE,
    june_3              DOUBLE,
    july_3              DOUBLE,
    august_3            DOUBLE,
    september_3         DOUBLE,
    october_3           DOUBLE,
    november_3          DOUBLE,
    december_3          DOUBLE,
    january_3           DOUBLE,
    february_3          DOUBLE,
    march_3             DOUBLE,
    april_3             DOUBLE,
    may_7               DOUBLE,
    june_4              DOUBLE,
    july_4              DOUBLE,
    august_4            DOUBLE,
    september_4         DOUBLE,
    october_4           DOUBLE,
    november_4          DOUBLE,
    december_4          DOUBLE,
    audit_etl_id        BIGINT      NOT NULL,
    audit_create_date   TIMESTAMP   NOT NULL,
    audit_update_date   TIMESTAMP   NOT NULL,
    audit_is_deleted    BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["BPC_Plug_Financials_MachineCosts"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.BPC_Plug_Financials_MachineCosts (
    unnamed_1           STRING,
    jan                 DOUBLE,
    feb                 DOUBLE,
    mar                 DOUBLE,
    apr                 DOUBLE,
    may                 DOUBLE,
    jun                 DOUBLE,
    jul                 DOUBLE,
    aug                 DOUBLE,
    sep                 DOUBLE,
    oct                 DOUBLE,
    nov                 DOUBLE,
    dec                 DOUBLE,
    jan_1               DOUBLE,
    feb_1               DOUBLE,
    mar_1               DOUBLE,
    apr_1               DOUBLE,
    may_1               DOUBLE,
    jun_1               DOUBLE,
    jul_1               DOUBLE,
    aug_1               DOUBLE,
    sep_1               DOUBLE,
    oct_1               DOUBLE,
    nov_1               DOUBLE,
    dec_1               DOUBLE,
    jan_2               DOUBLE,
    feb_2               DOUBLE,
    mar_2               DOUBLE,
    apr_2               DOUBLE,
    may_2               DOUBLE,
    jun_2               DOUBLE,
    jul_2               DOUBLE,
    aug_2               DOUBLE,
    sep_2               DOUBLE,
    oct_2               DOUBLE,
    nov_2               DOUBLE,
    dec_2               DOUBLE,
    jan_3               DOUBLE,
    feb_3               DOUBLE,
    mar_3               DOUBLE,
    apr_3               DOUBLE,
    may_3               DOUBLE,
    june                DOUBLE,
    july                DOUBLE,
    august              DOUBLE,
    september           DOUBLE,
    october             DOUBLE,
    november            DOUBLE,
    december            DOUBLE,
    january             DOUBLE,
    february            DOUBLE,
    march               DOUBLE,
    april               DOUBLE,
    may_4               DOUBLE,
    june_1              DOUBLE,
    july_1              DOUBLE,
    august_1            DOUBLE,
    september_1         DOUBLE,
    october_1           DOUBLE,
    november_1          DOUBLE,
    december_1          DOUBLE,
    january_1           DOUBLE,
    february_1          DOUBLE,
    march_1             DOUBLE,
    april_1             DOUBLE,
    may_5               DOUBLE,
    june_2              DOUBLE,
    july_2              DOUBLE,
    august_2            DOUBLE,
    september_2         DOUBLE,
    october_2           DOUBLE,
    november_2          DOUBLE,
    december_2          DOUBLE,
    january_2           DOUBLE,
    february_2          DOUBLE,
    march_2             DOUBLE,
    april_2             DOUBLE,
    may_6               DOUBLE,
    june_3              DOUBLE,
    july_3              DOUBLE,
    august_3            DOUBLE,
    september_3         DOUBLE,
    october_3           DOUBLE,
    november_3          DOUBLE,
    december_3          DOUBLE,
    january_3           DOUBLE,
    february_3          DOUBLE,
    march_3             DOUBLE,
    april_3             DOUBLE,
    may_7               DOUBLE,
    june_4              DOUBLE,
    july_4              DOUBLE,
    august_4            DOUBLE,
    september_4         DOUBLE,
    october_4           DOUBLE,
    november_4          DOUBLE,
    december_4          DOUBLE,
    audit_etl_id        BIGINT      NOT NULL,
    audit_create_date   TIMESTAMP   NOT NULL,
    audit_update_date   TIMESTAMP   NOT NULL,
    audit_is_deleted    BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["BW_COPA_DailyFeePoker_Units"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.BW_COPA_DailyFeePoker_Units (
    sales_org               BIGINT,
    juris_num               BIGINT,
    juris                   STRING,
    profit_ctr              BIGINT,
    sold_to_cust            BIGINT,
    sold_to_name            STRING,
    install_at              BIGINT,
    install_at_name         STRING,
    doc_type                STRING,
    doc_number              BIGINT,
    line_item               BIGINT,
    material_num            STRING,
    serial_num              STRING,
    theme                   STRING,
    theme_id                STRING,
    target_qty              BIGINT,
    from_date               DATE,
    to_date                 DATE,
    billing_date            DATE,
    no_of_billing_days      BIGINT,
    net_price               DOUBLE,
    locl_net_p              DOUBLE,
    net_value               DOUBLE,
    locl_net_v              DOUBLE,
    ag_theme                STRING,
    eom_installed_base      BIGINT,
    corp_cust_no            BIGINT,
    corp_cust_name          STRING,
    eom_machine_days        BIGINT,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["BW_ZRBINQ_BillingPlans"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.BW_ZRBINQ_BillingPlans (
    end_of_month_date       DATE,
    install_at              BIGINT,
    serial_num              STRING,
    coin_in                 DOUBLE,
    gross_win               DOUBLE,
    cust_split              DOUBLE,
    igt_split               DOUBLE,
    gross_igt_rev           DOUBLE,
    net_igt_rev             DOUBLE,
    machine_days            BIGINT,
    par_split               DOUBLE,
    type                    STRING,
    doc_type                STRING,
    doc_number              BIGINT,
    line_item               BIGINT,
    sold_to_cust            BIGINT,
    sold_to_name            STRING,
    install_at_name         STRING,
    profit_ctr              BIGINT,
    profit_ctr_desc         STRING,
    material_num            STRING,
    theme                   STRING,
    theme_id                STRING,
    short_text              STRING,
    bill_status             STRING,
    net_value               DOUBLE,
    net_price               DOUBLE,
    go_live                 DATE,
    juris                   STRING,
    acct_exec_name          STRING,
    customer_group          STRING,
    juris_num               BIGINT,
    sales_group             STRING,
    sales_office            BIGINT,
    sales_org               BIGINT,
    acct_exec               BIGINT,
    add_value_days          BIGINT,
    bill_plan               BIGINT,
    denom                   STRING,
    install                 DATE,
    period_len              BIGINT,
    system                  STRING,
    equipment_num           BIGINT,
    promo                   STRING,
    condition_type          STRING,
    no_of_billing_days      BIGINT,
    target_qty              BIGINT,
    from_date               DATE,
    to_date                 DATE,
    billing_date            DATE,
    fiscal_date             DATE,
    orig_doc                STRING,
    orig_item               STRING,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["GameOpsInstallBaseActivity"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.GameOpsInstallBaseActivity (
    date_to                 DATE,
    bingo_non_premium       STRING,
    serial_num              STRING,
    customer_number         BIGINT,
    system_code             BIGINT,
    brand_code              BIGINT,
    theme_code_updated      STRING,
    unit_type               STRING,
    activity                STRING,
    stl_flag                STRING,
    material_updated        STRING,
    profit_center           BIGINT,
    install_date            DATE,
    go_live_date            DATE,
    shutdown_date           DATE,
    removal_date            DATE,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["GameOpsInstallBasePSD"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.GameOpsInstallBasePSD (
    date_to                 DATE,
    product_type            STRING,
    serial_num              STRING,
    customer_number         BIGINT,
    system_code             BIGINT,
    brand_code              BIGINT,
    theme_code_fix          STRING,
    unit_type               STRING,
    material_updated        STRING,
    stl_flag_cosmo          STRING,
    profit_center           BIGINT,
    install_date            DATE,
    go_live_date            DATE,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Exp_Depreciation"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Exp_Depreciation (
    co_code                     BIGINT,
    asset                       BIGINT,
    sub                         BIGINT,
    asset_sub                   STRING,
    class                       STRING,
    life                        STRING,
    serialnum                   STRING,
    material                    STRING,
    asset_description           STRING,
    group                       STRING,
    group_2                     STRING,
    nvglacct                    BIGINT,
    nvadglacct                  BIGINT,
    nvdeprexpgl                 BIGINT,
    cfinglacct                  BIGINT,
    cfinadglacct                BIGINT,
    cfin_fs_rollup              STRING,
    inst_at                     BIGINT,
    customer_name               STRING,
    status                      STRING,
    cost_ctr                    BIGINT,
    cost_ctr_description        STRING,
    cost_center_owner           STRING,
    jurisdiction                STRING,
    func_area                   STRING,
    sys_category                STRING,
    plca_group                  STRING,
    global_plca_name            STRING,
    brand                       STRING,
    game_theme                  STRING,
    capdate                     DATE,
    depr_start_date             DATE,
    depky                       STRING,
    capval_usd                  DOUBLE,
    accum_val_usd               DOUBLE,
    netbook_usd                 DOUBLE,
    salvage_value               DOUBLE,
    january                     DOUBLE,
    february                    DOUBLE,
    march                       DOUBLE,
    april                       DOUBLE,
    may                         DOUBLE,
    june                        DOUBLE,
    july                        DOUBLE,
    august                      DOUBLE,
    september                   DOUBLE,
    october                     DOUBLE,
    november                    DOUBLE,
    december                    DOUBLE,
    total                       DOUBLE,
    audit_etl_id                BIGINT      NOT NULL,
    audit_create_date           TIMESTAMP   NOT NULL,
    audit_update_date           TIMESTAMP   NOT NULL,
    audit_is_deleted            BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Intl_Performance_EMEA_Africa"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Intl_Performance_EMEA_Africa (
    serial_num                  STRING,
    doc_number                  BIGINT,
    line_item                   BIGINT,
    doc_type                    STRING,
    equipment_num               BIGINT,
    sold_to_cust                BIGINT,
    sold_to_name                STRING,
    material_num                STRING,
    theme                       STRING,
    target_qty                  BIGINT,
    from_date                   DATE,
    to_date                     DATE,
    billing_date                DATE,
    net_price                   DOUBLE,
    no_of_billing_days          BIGINT,
    net_value                   DOUBLE,
    brand                       STRING,
    profit_ctr_desc             STRING,
    profit_ctr                  BIGINT,
    bill_status                 STRING,
    theme_id                    STRING,
    install_at                  BIGINT,
    install_at_name             STRING,
    condition_type              STRING,
    acct_exec                   BIGINT,
    acct_exec_name              STRING,
    sales_org                   BIGINT,
    sales_office                BIGINT,
    sales_group                 STRING,
    bill_plan                   BIGINT,
    short_text                  STRING,
    period_len                  BIGINT,
    add_value_days              BIGINT,
    local_net_value             DOUBLE,
    local_net_price             DOUBLE,
    system_id                   BIGINT,
    system                      STRING,
    brand_id                    BIGINT,
    promo                       STRING,
    denom                       STRING,
    juris_num                   BIGINT,
    juris                       STRING,
    install                     DATE,
    go_live                     DATE,
    fiscal_date                 DATE,
    customer_group              STRING,
    original_document           BIGINT,
    original_document_item      BIGINT,
    audit_etl_id                BIGINT      NOT NULL,
    audit_create_date           TIMESTAMP   NOT NULL,
    audit_update_date           TIMESTAMP   NOT NULL,
    audit_is_deleted            BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Intl_Performance_EMEA_FixedFee"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Intl_Performance_EMEA_FixedFee (
    premiumcategory             STRING,
    recurring_type              STRING,
    unitgrouping                STRING,
    mlp                         STRING,
    serial_number               STRING,
    sold_to_number              BIGINT,
    install_at_number           BIGINT,
    sold_to_name                STRING,
    install_at_name             STRING,
    group                       STRING,
    country                     STRING,
    jurisdiction                STRING,
    region                      STRING,
    pc_country                  STRING,
    profit_center               BIGINT,
    profit_center_description   STRING,
    sap_material                STRING,
    cabinet                     STRING,
    brand_description           STRING,
    theme_description           STRING,
    pricing_local_currency_pmpm DOUBLE,
    local_currency              STRING,
    pricing_usd_pdpm            DOUBLE,
    sales_teamleader            STRING,
    account_manager             STRING,
    casino_opening_date_mm_dd_yyyy_enter_day_first  DATE,
    unit_status                 STRING,
    invoice_amount_usd          DOUBLE,
    invoice_amount_local_currency   DOUBLE,
    total_active_days           BIGINT,
    total_in_active_days        BIGINT,
    check                       STRING,
    total_cn_value              DOUBLE,
    notes_changes_in_fees_any_other_comments    STRING,
    correct_invoice_amount_usd_for_the_last_month_where_corrections_are_required    DOUBLE,
    correct_credit_amount_usd_for_the_last_month_where_corrections_are_required     DOUBLE,
    total_revenue_usd           DOUBLE,
    go_live_date_mm_dd_yyyy_enter_day_first     DATE,
    future_date_for_update_daily_fee_removal_option_to_terminate_contract_change_etc_mm_dd_yyyy DATE,
    buyout_option_enter_date_mm_dd_yyyy_or_no   STRING,
    opp_number                  STRING,
    quote_number                BIGINT,
    docs_in_salesforce          STRING,
    opp_name                    STRING,
    salesforce_product          STRING,
    contract_expiration_date_mm_dd_yyyy_enter_day_first DATE,
    new_unit_type_yes_also_when_you_entered_yes_in_column_novation_otherwise_it_stays_blank STRING,
    audit_etl_id                BIGINT      NOT NULL,
    audit_create_date           TIMESTAMP   NOT NULL,
    audit_update_date           TIMESTAMP   NOT NULL,
    audit_is_deleted            BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Intl_Performance_EMEA_GreeceWLA"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Intl_Performance_EMEA_GreeceWLA (
    month_of_period_start           DATE,
    machine_id_serial               BIGINT,
    sum_of_coin_in                  DOUBLE,
    sum_of_net_win                  DOUBLE,
    net_win_2_players_reward_deducted   DOUBLE,
    egm_day_count                   BIGINT,
    wpupd                           DOUBLE,
    fee_revenue                     DOUBLE,
    distinct_ser_count              BIGINT,
    audit_etl_id                    BIGINT      NOT NULL,
    audit_create_date               TIMESTAMP   NOT NULL,
    audit_update_date               TIMESTAMP   NOT NULL,
    audit_is_deleted                BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Intl_Performance_LAC"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Intl_Performance_LAC (
    serial_number           STRING,
    posting_period          DATE,
    sold_to_cust_num        BIGINT,
    source_currency         STRING,
    gross_win               DOUBLE,
    net_win                 DOUBLE,
    igt_split               DOUBLE,
    discount                DOUBLE,
    net_igt_rev             DOUBLE,
    days_on_line            BIGINT,
    country                 STRING,
    brand                   STRING,
    theme                   STRING,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Peformance_PremLotto_DE"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Peformance_PremLotto_DE (
    track                   STRING,
    vendor                  STRING,
    theme                   STRING,
    license                 BIGINT,
    amount_played           DOUBLE,
    amount_won              DOUBLE,
    net_revenue             DOUBLE,
    _                       STRING,
    igt_fee                 DOUBLE,
    startdate               DATE,
    enddate                 DATE,
    pc                      BIGINT,
    profit_ctr_desc         STRING,
    brand                   STRING,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Peformance_PremLotto_NY"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Peformance_PremLotto_NY (
    location                STRING,
    devicename              STRING,
    igt_sn                  BIGINT,
    profit_ctr              BIGINT,
    profit_ctr_desc         STRING,
    brand                   STRING,
    credits_played          BIGINT,
    coupons_played          DOUBLE,
    credits_won             BIGINT,
    net_win_credits         DOUBLE,
    fee_7_25                DOUBLE,
    theme                   STRING,
    period                  STRING,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Peformance_PremLotto_RI"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Peformance_PremLotto_RI (
    date                    DATE,
    vlt_desc                STRING,
    serial_number           STRING,
    cash_in                 DOUBLE,
    cash_out                DOUBLE,
    net_income              DOUBLE,
    commission              DOUBLE,
    igt_fee                 DOUBLE,
    location                STRING,
    vendor                  STRING,
    game_set                STRING,
    pc                      BIGINT,
    pc_desc                 STRING,
    note                    STRING,
    audit_etl_id            BIGINT      NOT NULL,
    audit_create_date       TIMESTAMP   NOT NULL,
    audit_update_date       TIMESTAMP   NOT NULL,
    audit_is_deleted        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Performance_CDS_Class_II"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Performance_CDS_Class_II (
    install_at                  BIGINT,
    document_num                BIGINT,
    invoice_date_from           DATE,
    invoice_date_to             DATE,
    jurisdiction                STRING,
    mjpjur                      STRING,
    mjsegment                   STRING,
    install_at_cust_name        STRING,
    serial_num                  STRING,
    brand_code                  STRING,
    brand                       STRING,
    theme_code                  STRING,
    theme                       STRING,
    install_at_state            STRING,
    aggregated_amount_wagered   DOUBLE,
    aggregated_amount_won       DOUBLE,
    gross_win                   DOUBLE,
    aggregated_machine_days     BIGINT,
    lease_rate                  DOUBLE,
    flat_rate                   DOUBLE,
    min_per_day                 DOUBLE,
    casino_split                DOUBLE,
    igt_split                   DOUBLE,
    total_rev                   DOUBLE,
    total_discount              DOUBLE,
    net_amount                  DOUBLE,
    concession_amount           DOUBLE,
    audit_etl_id                BIGINT      NOT NULL,
    audit_create_date           TIMESTAMP   NOT NULL,
    audit_update_date           TIMESTAMP   NOT NULL,
    audit_is_deleted            BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Performance_CDS_Class_II_ManualBilling"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Performance_CDS_Class_II_ManualBilling (
    end_of_month_date                       DATE,
    install_at                              BIGINT,
    serial_num                              STRING,
    coin_in                                 DOUBLE,
    gross_win                               DOUBLE,
    cust_split                              DOUBLE,
    igt_split                               DOUBLE,
    gross_igt_rev                           DOUBLE,
    net_igt_rev                             DOUBLE,
    machine_days                            BIGINT,
    par_split                               DOUBLE,
    cust_name_ignore_for_sql_upload         STRING,
    audit_etl_id                            BIGINT      NOT NULL,
    audit_create_date                       TIMESTAMP   NOT NULL,
    audit_update_date                       TIMESTAMP   NOT NULL,
    audit_is_deleted                        BOOLEAN     NOT NULL
) USING DELTA
"""

SILVER_DDL["RPM_Performance_WAP_NewCanada"] = """
CREATE TABLE IF NOT EXISTS {env}_022_silver_finance.atlas_map.RPM_Performance_WAP_NewCanada (
    brand_code                  STRING,
    theme_code                  STRING,
    system                      STRING,
    customer_num_install_at     BIGINT,
    serial_num                  STRING,
    eod_date                    DATE,
    machine_hours               BIGINT,
    meter_wagered_amount        DOUBLE,
    ceje_amount                 DOUBLE,
    contribution_rate           DOUBLE,
    wap_operator_fee            DOUBLE,
    daily_cap_setting_amount    DOUBLE,
    daily_cap_disc_amount       DOUBLE,
    daily_min_setting_amount    DOUBLE,
    daily_min_disc_amount       DOUBLE,
    oos_discount                DOUBLE,
    daily_flat_fee              DOUBLE,
    wap_operator_fee_discount   DOUBLE,
    total_operator_fee          DOUBLE,
    final_contribution_amount   DOUBLE,
    audit_etl_id                BIGINT      NOT NULL,
    audit_create_date           TIMESTAMP   NOT NULL,
    audit_update_date           TIMESTAMP   NOT NULL,
    audit_is_deleted            BOOLEAN     NOT NULL
) USING DELTA
"""

def cast_to_silver(df, fields):
    type_map = {f.name.lower(): f.dataType for f in fields}
    col_map  = {c.lower(): c for c in df.columns}
    exprs = []
    for col, dtype in type_map.items():
        if col not in col_map:
            continue
        src = col_map[col]
        if isinstance(dtype, (T.LongType, T.IntegerType, T.ShortType)):
            exprs.append(F.col(src).cast("double").cast("long").alias(col))
        elif isinstance(dtype, (T.DoubleType, T.FloatType, T.DecimalType)):
            exprs.append(F.col(src).cast("double").alias(col))
        elif isinstance(dtype, T.DateType):
            exprs.append(F.col(src).cast("date").alias(col))
        elif isinstance(dtype, T.TimestampType):
            exprs.append(F.col(src).cast("timestamp").alias(col))
        elif isinstance(dtype, T.BooleanType):
            exprs.append(F.col(src).cast("boolean").alias(col))
        else:
            exprs.append(F.col(src).cast("string").alias(col))
    return df.select(exprs)
```

```python
try:
    env = dbutils.secrets.get(scope="workspace-details", key="name")
except Exception:
    env = "development"

bronze = f"{env}_021_bronze_finance.atlas"
stage  = f"{env}_022_silver_finance.atlas_stage"
silver = f"{env}_022_silver_finance.atlas_map"

run_id = getRunDetails()["jobRunId"]
rows   = spark.sql(f"SELECT etl_id FROM {env}_011_bronze_core.db_admin.pipeline_history WHERE job_run_id = '{run_id}'").collect()
etl_id = rows[0]["etl_id"] if rows else None

spark.sql(f"CREATE SCHEMA IF NOT EXISTS {env}_022_silver_finance.atlas_stage")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {env}_022_silver_finance.atlas_map")

print(f"env    : {env}")
print(f"bronze : {bronze}")
print(f"stage  : {stage}")
print(f"silver : {silver}")
print(f"etl_id : {etl_id}")
print()

total   = len(ATLAS_TABLES)
results = []

for i, table in enumerate(ATLAS_TABLES, 1):
    src_bronze = f"{bronze}.{table}"
    src_stage  = f"{stage}.{table}"
    tgt_silver = f"{silver}.{table}"

    print(f"[{i}/{total}] {table}")

    try:
        if spark.catalog.tableExists(src_stage):
            df     = spark.sql(f"SELECT * FROM {src_stage}")
            source = "stage"
        else:
            df     = spark.sql(f"SELECT * FROM {src_bronze} WHERE IS_DELETED = false")
            source = "bronze"

        drop_cols = {"etl_id", "file_id", "is_deleted", f"{table.lower()}_id"}
        col_map   = {c.lower(): c for c in df.columns}
        df        = df.drop(*[col_map[c] for c in drop_cols if c in col_map])

        if not spark.catalog.tableExists(tgt_silver):
            spark.sql(SILVER_DDL[table].replace("{env}", env))
            print(f"  -> created silver table")

        audit_cols = {"audit_etl_id", "audit_create_date", "audit_update_date", "audit_is_deleted"}
        schema     = spark.read.table(tgt_silver).schema
        fields     = [f for f in schema.fields if f.name.lower() not in audit_cols]
        df         = cast_to_silver(df, fields)

        now = F.current_timestamp()
        df  = (df
            .withColumn("audit_etl_id",     F.lit(etl_id).cast("long"))
            .withColumn("audit_create_date", now)
            .withColumn("audit_update_date", now)
            .withColumn("audit_is_deleted",  F.lit(False))
        )

        n = df.count()
        spark.sql(f"TRUNCATE TABLE {tgt_silver}")
        df.write.format("delta").mode("append").saveAsTable(tgt_silver)

        print(f"  -> {n:,} rows loaded from {source}")
        results.append((table, source, n, "OK"))

    except Exception as e:
        print(f"  -> ERROR: {e}")
        results.append((table, "?", 0, f"ERROR: {e}"))

print()
print(f"Done — {total} tables processed")
print()

ok  = [r for r in results if r[3] == "OK"]
err = [r for r in results if r[3] != "OK"]

for name, src, n, _ in ok:
    print(f"  OK   {name:50s}  {n:>8,} rows  [{src}]")

if err:
    print()
    for name, src, n, status in err:
        print(f"  FAIL {name:50s}  {status}")
    raise Exception(f"{len(err)} table(s) failed")
```
